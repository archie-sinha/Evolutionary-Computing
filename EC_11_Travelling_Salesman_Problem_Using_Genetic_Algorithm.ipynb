###1. Travelling Salesman Problem (TSP):

- Problem statement-
    The Traveling Salesman Problem (TSP) is about finding the shortest route that visits each city once and returns to the starting city. It's like a salesman who wants to visit every city on his list, but he wants to do it in the shortest distance possible.
- How do you solve the Travelling Salesman Problem (TSP) generally?
   
    Here are different ways to solve TSP:

    a. Exact Algorithms: These find the best solution but can be slow for many cities.

    b. Heuristic Algorithms: These find good solutions quickly but might not be the best.

    c. Metaheuristic Algorithms: These are smart strategies that try to find good solutions by exploring different routes.

###2. Genetic Algorithm:

- An evolutionary approach to solving the Traveling Salesman Problem (TSP) involves representing candidate solutions as complete tours, where each tour specifies the order in which cities are visited.

- Here's how a genetic algorithm (GA) can be applied to solve the TSP:
    1. Initialization:

      Generate an initial population of candidate solutions. Each solution represents a complete tour that visits each city exactly once and returns to the starting city.
      Randomly generate k tours, where each tour is a permutation of the cities.

    2. Evaluation:

    Evaluate the fitness of each tour in the population. The fitness is typically determined by the total distance travelled along the tour. Shorter tours have higher fitness.

    3. Selection:

    Select a subset of tours from the population for reproduction based on their fitness. Higher fitness tours have a higher chance of being selected.
    Common selection methods include roulette wheel, tournament, and rank-based selection.

    4. Crossover:

    Perform crossover (recombination) between selected tours to create offspring. Crossover involves combining two parent tours to produce new tours (offspring) with characteristics of both parents.
    Common crossover techniques for TSP include partially mapped crossover (PMX), order crossover (OX), and cycle crossover (CX).

    5. Mutation:

    Apply mutation to introduce genetic diversity into the population. Mutation involves making small random changes to the tours to explore new areas of the solution space.
    Common mutation operators for TSP include swap mutation, insertion mutation, and inversion mutation.

    6. Replacement:

    Replace some tours in the current population with offspring generated from crossover and mutation operations. This ensures that the population size remains constant.

    7. Termination:

    Repeat steps 2 to 6 for a predefined number of generations or until a termination condition is met (e.g., reaching a maximum number of iterations, finding a satisfactory solution).

    8. Solution Extraction:
    
    Extract the best tour (solution) from the final population. The best tour represents the optimal or near-optimal solution to the TSP.

##Task:
1. Create a population of candidate paths (tours) representing salesman routes.
2. Evaluate each path's fitness based on total distance (shorter = fitter).
3. Select parents based on fitness (e.g., roulette wheel selection).
4. Apply crossover (e.g., single-point, order crossover) to combine parent paths.
5. Introduce mutations (e.g., swapping cities) for variation.
6. Repeat selection, crossover, and mutation to generate new generations.
7. Observe how the algorithm converges to an optimal or near-optimal route.


```python
import random
import numpy as np
```

- Create a population of candidate paths (tours) representing salesman routes.

```python
# Define the distances between cities
distances = [
    [0, 10, 15, 20, 25],
    [10, 0, 35, 25, 30],
    [15, 35, 0, 30, 10],
    [20, 25, 30, 0, 40],
    [25, 30, 10, 40, 0]
]
```

```python
# Number of cities
num_cities = 5
```

```python
# Population size
population_size = 10
```

```python
# Number of generations
num_generations = 100
```

- Evaluate each path's fitness based on total distance (shorter = fitter).


```python
# Function to calculate total distance of a path
def calculate_total_distance(path):
    total_distance = 0
    for i in range(num_cities - 1):
        total_distance += distances[path[i]][path[i + 1]]
    total_distance += distances[path[num_cities - 1]][path[0]]
    return total_distance
```

```python
# Initialize population with random paths
population = [random.sample(range(num_cities), num_cities) for _ in range(population_size)]
```

- Select parents based on fitness (e.g., roulette wheel selection).
- Apply crossover (e.g., single-point, order crossover) to combine parent paths.
- Introduce mutations (e.g., swapping cities) for variation.
- Repeat selection, crossover, and mutation to generate new generations.


```python
# Main Genetic Algorithm loop
for generation in range(num_generations):
    # Evaluate fitness of each path
    fitness_scores = [1 / calculate_total_distance(path) for path in population]

    # Select parents based on fitness using roulette wheel selection
    parents = random.choices(population, weights=fitness_scores, k=2)

    # Apply crossover (single-point crossover)
    crossover_point = random.randint(1, num_cities - 1)
    child = parents[0][:crossover_point] + [city for city in parents[1] if city not in parents[0][:crossover_point]]

    # Introduce mutation (swap two cities)
    if random.random() < 0.1:
        swap_indices = random.sample(range(num_cities), 2)
        child[swap_indices[0]], child[swap_indices[1]] = child[swap_indices[1]], child[swap_indices[0]]

    # Replace the least fit path with the new child
    worst_path_index = fitness_scores.index(min(fitness_scores))
    population[worst_path_index] = child
```

```python
# Find the best path after all generations
best_path_index = fitness_scores.index(max(fitness_scores))
best_path = population[best_path_index]
best_distance = calculate_total_distance(best_path)
```

- Observe how the algorithm converges to an optimal or near-optimal route.

```python
print("Best Path:", best_path)
print("Best Distance:", best_distance)
```
