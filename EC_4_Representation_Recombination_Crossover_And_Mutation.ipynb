###**Representation**
- There are different ways in which solutions can be represented and varied within the overall search algorithm.
- The first stage of building any evolutionary algorithm is to decide on a genetic representation of a candidate solution to the problem.
- This involves defining the genotype and the mapping from genotype to phenotype.
- When choosing a representation, it is important to choose the right representation for the problem being solved.
- Choosing the right representation only comes with practice and a good knowledge of the application domain.

*Case 1: In a supply chain management problem, a chromosome typically represents a potential solution to optimize the supply chain. The genes in the chromosome could represent various aspects of the supply chain, such as suppliers, warehouses, transportation routes, etc. Generate a chromosome where each gene represents the index of the supplier for a product. the gene represents the supplier attached*

1. **Representation**: Integer- represents the index of the supplier for a product
2. **Recombination/Crossover**: One-point - random crossover point is selected, and the genes before the crossover point are swapped between the parents to create two new offspring
3. **Mutation**: Random mutation - a random gene in the chromosome is selected, and its value is randomly changed to another valid value

```python
#Integer Representation
chromosome_1= [2,1,3,4,6,5]     #Parent Chromosome 1
chromosome_2= [3,1,2,7,4,5]     #Parent Chromosome 2
```

```python
#One-point Recombination
r_point=2   #random selection of crossover point

offspring_1=[]
offspring_2=[]

def onepoint_crossover(chromosome_1, chromosome_2, point):
    offspring_1 = chromosome_1[:point] + chromosome_2[point:]
    offspring_2 = chromosome_2[:point] + chromosome_1[point:]
    return offspring_1, offspring_2
```

```python
# print(offspring_1)
# print(offspring_2)
Crossover= onepoint_crossover(chromosome_1, chromosome_2, r_point)
print(Crossover)
```

```python
#Random Mutation
m_point=4   #random selection of mutation point

offspring_3=[]
offspring_4=[]

import random

def random_mutation(chromosome_1, chromosome_2, m_point):


  random_int_1 = random.randint(0, 9)
  random_int_2= random .randint(0,9)

  offspring_3= chromosome_1[0:(m_point-1)]
  offspring_3.append(random_int_1)
  offspring_3.append(chromosome_1[(m_point):])

  offspring_4= chromosome_2[0:(m_point-1)]
  offspring_4.append(random_int_2)
  offspring_4.append(chromosome_2[(m_point):])

  return offspring_3, offspring_4
```

```python
# print(offspring_3)
# print(offspring_4)
Mutation= random_mutation(chromosome_1, chromosome_2, m_point)
print(Mutation)
```

*Case 2: In the context of evaluating students' performance, a chromosome might represent a possible combination of grades or scores for each student or each subject. Consider that each gene in the chromosome represents the score of a student in a particular subject*
1. **Representation**: Integer - each gene represents the score of a student in a particular subject
2. **Recombination**: Two-point - two random crossover points are selected, and the genes between the crossover points are swapped between the parents to create two new offspring
3. **Mutation**: Random mutation - a random gene in the chromosome is selected, and its value is randomly changed to another valid value

```python
#Integer Representation
chromosome_1= [2,1,3,4,6,5]     #Parent Chromosome 1
chromosome_2= [3,4,2,7,4,5]     #Parent Chromosome 2
```

```python
#Two-point Recombination
r_point_1=2   #random selection of crossover point
r_point_2=4

offspring_1=[]
offspring_2=[]

def twopoint_crossover(chromosome_1, chromosome_2, r_point_1, r_point_2):
    offspring_1 = chromosome_1[:r_point_1] + chromosome_2[r_point_1:r_point_2] + chromosome_1[r_point_2:]
    offspring_2 = chromosome_2[:r_point_1] + chromosome_1[r_point_1:r_point_2] + chromosome_2[r_point_2:]
    return offspring_1, offspring_2
```

```python
# print(offspring_1)
# print(offspring_2)
Crossover= twopoint_crossover(chromosome_1, chromosome_2, r_point_1, r_point_2)
print(Crossover)
```

*Case 3: Chromosome representation for Fractional Knapsack Problem (in Fractional Knapsack, the items can be chosen in fractions as well to be placed in the knapsack)*

1. **Representation**: Real Valued / Float- each gene represents the fraction of an item to be placed in the knapsack
2. **Recombination/Crossover**: Uniform crossover - each gene in the offspring is randomly selected from one of the parents
3. **Mutation**: Random mutation - a random gene in the chromosome is selected, and its value is randomly changed to another valid value

```python
#Real Valued / Float Representation
chromosome_1= [0.2,0.1,0.3,0.4,0.6,0.5]     #Parent Chromosome 1
chromosome_2= [0.3,0.4,0.2,0.7,0.4,0.5]     #Parent Chromosome 2
```

```python
# Uniform Crossover
offspring_1 = []
offspring_2 = []
def uniform_crossover(chromosome_1, chromosome_2, probability=0.5):

    for gene_1, gene_2 in zip(chromosome_1, chromosome_2):
        if random.uniform(0, 1) < probability:
            offspring_1.append(gene_1)
            offspring_2.append(gene_2)
        else:
            offspring_1.append(gene_2)
            offspring_2.append(gene_1)

    return offspring_1, offspring_2
```

```python
# print(offspring_1)
# print(offspring_2)
Crossover= uniform_crossover(chromosome_1, chromosome_2)
print(Crossover)
```

```python
#Random Mutation
m_point=4   #random selection of mutation point

offspring_3=[]
offspring_4=[]

import random

def random_mutation(chromosome_1, chromosome_2, m_point):


  random_int_1 = random.randint(0, 9)
  random_int_2= random .randint(0,9)

  offspring_3= chromosome_1[0:(m_point-1)]
  offspring_3.append(random_int_1)
  offspring_3.append(chromosome_1[(m_point):])

  offspring_4= chromosome_2[0:(m_point-1)]
  offspring_4.append(random_int_2)
  offspring_4.append(chromosome_2[(m_point):])

  return offspring_3, offspring_4
```

##**Conclusion:**
Representation: It refers to how individuals in a genetic algorithm are encoded, such as using binary strings or numerical arrays to represent solutions.

Crossover: A genetic operator in which segments of genetic material from two parents are combined to produce new offspring, influencing the exploration of the solution space.

Mutation: A genetic operator that introduces random changes in the genetic material of an individual, promoting diversity and preventing premature convergence in genetic algorithms.

##**Curiosity Questions**

1. How does the mutation rate impact the convergence and diversity of the population in genetic algorithm.

- Convergence: A higher mutation rate increases exploration, helping the algorithm escape local optima. However, too high a rate may hinder convergence and slow down the convergence towards an optimal solution.
- Diversity: A lower mutation rate may lead to premature convergence, with the population converging to a suboptimal solution. A balanced mutation rate is crucial for maintaining diversity and exploring the solution space effectively.


2. What considerations should be taken into account when selecting a crossover operator for a specific problem.

- Problem Characteristics: Different problems may benefit from specific crossover operators. For example, permutation-based problems might require a different crossover than numerical optimization problems.
- Exploration vs. Exploitation: Uniform crossover promotes exploration, while one-point or two-point crossover emphasizes exploitation. The choice depends on whether a balance of exploration and exploitation is needed.
- Solution Representation: The representation of solutions (binary, real-valued, permutation) influences the suitability of crossover operators. Choose an operator compatible with the representation for effective information exchange.
- Preservation of Building Blocks: Some crossover operators better preserve essential solution components or building blocks. Tailor the choice to the problem's characteristics to maintain useful genetic material.
